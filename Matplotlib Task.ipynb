### 5. a. Create a figure object and add two axes - axes1 and axes2. Pass the following lists as parameters - [0, 0, 1, 1] to axes1 and [0.2, 0.5, 0.3, 0.4] to axes2
### b. Create plots using a, b, c arrays to match the image below. Add limits to the plots - x limit equals (20, 60) and y limit equals (10, 70) for axes1 and x limit equals (20, 60) and y limit equals (10000, 70000) for axes2

```python
figure= plt.figure()
axes1= figure.add_axes([0,0,1,1])
axes2= figure.add_axes([0.2, 0.5, 0.3, 0.4])

axes1.plot(a, b)
axes1.set_xlabel('a array')
axes1.set_ylabel('b array')
axes1.set_title('Outer plot')
axes1.set_xlim(20, 60)
axes1.set_ylim(10, 70)

axes2.plot(a, c)
axes2.set_xlabel('a array')
axes2.set_ylabel('b array')
axes2.set_title('Inner plot')
axes2.set_xlim(20, 60)
axes2.set_ylim(10000, 70000)
```

### 6. a. Create a figure object with 2 axes in 1 row with the help of subplots() method 
### b. Resize both axes - set 11 inches to width and 3 inches to height
### c. Create the plot on the first axes with c as x axis data and a as y axis data
### d. Create the plot on the second axes with c as x axis data and b as y axis data
### d. Change the appearance of plot lines somehow. Play with colors, width and styles

```python
figure, axes = plt.subplots(nrows= 1, ncols= 2, figsize=(11, 3))
axes[0].plot(c, a, color='purple', alpha=0.3, linestyle='--', lw=2, marker='<')
axes[1].plot(c, b, color='orange', linestyle=':', lw=5, marker='>')
```
